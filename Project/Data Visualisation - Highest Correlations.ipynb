## Plotting the highest correlation courses pre-post 2014

### Pre 2014

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

def to_strptime(string):
    try:
        return datetime.strptime(string, "%Y-%m-%d")
    except Exception as e:
        print(e)

os.chdir('C:\\Users\\yeachan153\\Desktop\\Methodology-Consulting\\Project')
data = pd.read_csv('Master Dataframe.csv')

data['Start Date'] = data['Start Date'].apply(to_strptime)

filter_14 = datetime.strptime('2014-09-01', "%Y-%m-%d")

pre_data = data[data['Start Date'] < filter_14]
post_data = data[data['Start Date'] >= filter_14]
```

```python
one = post_data[post_data['Adviesvaardigheden'].notnull()]
two = post_data[post_data['Masterstage A&O'].notnull()]
three = post_data[post_data['MAPS'].notnull()]
four = post_data[post_data['Pros. gedrag in werkrelat.'].notnull()]

list_df = [one,two,three,four]
```

```python
title2 = ['Adviesvaardigheden Vs. Thesis Grades', 'Masterstage.A.O Vs. Thesis Grades', 'MAPS Vs. Thesis Grades',
         'Pros. gedrag in werkrelat. Vs. Thesis Grades']

courses = ['Adviesvaardigheden','Masterstage A&O','MAPS','Pros. gedrag in werkrelat.']

text = ['rs = 0.534, p < 0.002, n = 35', 'rs = 0.492, p < 0.001, n = 74', 'rs = 0.481, p = 0.007, n = 33',
       'rs = 0.472, p = 0.013, n = 27']

fig = plt.figure(figsize = (55,35))
for i in range(4):
    ax = fig.add_subplot(2,2,i+1)
    sns.regplot(list_df[i][courses[i]], list_df[i]['thesis_grades'], color = 'black', y_jitter = 0.05)
    sns.set_style('darkgrid')
    ax.set_title(title2[i], fontsize = 65)
    plt.xticks(fontsize = 35)
    plt.yticks(fontsize = 35)
    plt.xlabel('Grade on ' + courses[i], fontsize = 55)
    plt.ylabel('Thesis Grade', fontsize = 55)
    ax.set_ylim(5.3,10)
    ax.set_xlim(6,10)
    if i == 0:
        ax.text(8.6, 5.5, text[i], fontsize = 35)
    if i == 1:
        ax.text(8.6,5.5, text[i], fontsize = 35)
    if i == 2:
        ax.text(8.6, 5.5, text[i], fontsize = 35)
    if i == 3:
        ax.text(8.6, 5.5, text[i], fontsize = 35)
```
